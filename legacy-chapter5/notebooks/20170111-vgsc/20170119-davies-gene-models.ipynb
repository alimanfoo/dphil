```python
def locate_agamp3_vgsc(query_seq):
    """Locate a query sequence within the AgamP3 genomic sequence for the Vgsc gene."""
    query_seq = query_seq.upper()
    for i in range(len(vgsc_seq_agamp3) - len(query_seq) + 1):
        comp_seq = vgsc_seq_agamp3[i:i+len(query_seq)].upper()
        if query_seq == comp_seq:
            return i + region_vgsc[1], i + region_vgsc[1] + len(query_seq) - 1
    return None
```

```python
# need to fix genomic coordinates, assume underlying sequence is same but coordinate system different
tbl_davies_liftover = (
    tbl_davies
    .addfield('ID', lambda row: '%s%s' % (row.Exon, row.Region))
    .split(1, '-', ['start', 'end'])
    .convert(['Exon', 'length', 'start', 'end'], int)
    .convert(['start', 'end'], lambda v: 1672917 - v + 2358158)
    .rename({'start': 'end', 'end': 'start'})
    .movefield('start', 15)
    .split(1, '-', ['mrna_start', 'mrna_end'])
    .convert(['mrna_start', 'mrna_end'], int)
    .addfield('seqid', '2L', index=1)
    # manual corrections
    .update('mrna_start', 1645, where=lambda row: row.Exon == 13)
    .update('mrna_end', 1707, where=lambda row: row.Exon == 13)
    .update('end', 2417799, where=lambda row: row.Region == 'c')
    .update('start', 2425770, where=lambda row: row.Region == 'k')
    .update('end', 2425892, where=lambda row: row.Region == 'k')
    # checks
    .addfield('agamp3_coords', lambda row: locate_agamp3_vgsc(davies_seq[row.mrna_start - 1:row.mrna_end]))
    .addfield('mrna_length_check', lambda row: row['length'] == (row.end - row.start + 1) == (row.mrna_end - row.mrna_start + 1))
    .addfield('coord_check', lambda row: row.agamp3_coords and row.start == row.agamp3_coords[0] and row.end == row.agamp3_coords[1])
    .cache()
)
tbl_davies_liftover.displayall(tr_style=lambda row: 'background-color: yellow' if not row.coord_check else 'background-color: pink' if not row.mrna_length_check else '')
```

```python
def pair_seq_comp(a, b):
    h1 = ''
    h2 = ''
    for x, y in itertools.zip_longest(a, b):
        if x == y:
            s = '<span style="background-color: black; color: white">%s</span>' % x
            h1 += s
            h2 += s
        else:
            if x:
                h1 += x
            if y:
                h2 += y
    from IPython.display import display_html
    html = '<pre>' + h1 + '\n' + h2 + '</pre>'
    display_html(html, raw=True)
```

## Build a GFF for the Davies transcripts

```python
davies_transcript_ids = 'C1N2 C3N2 C5N2 C7N2 C8N2 C10N2 C11N2 C1N9 C8N9 C1N9ck'.split()
```

```python
from urllib.parse import quote
```

```python
tbl = ['seqid source type start end score strand phase ID Parent'.split()]
# add the transcript records
for transcript_id in davies_transcript_ids:
    f = ['2L', 'Davies et al. (2007)', 'mRNA', region_vgsc[1], region_vgsc[2], '.', '+', '.', 'Davies-%s' % transcript_id, 'AGAP004707']
    tbl += [f]
# add the exons and CDSs
transcript_length = 0
for row in tbl_davies_liftover.records():
    for transcript_id in davies_transcript_ids:
        if row[transcript_id] == 'O':
            # include exon
            f = ['2L', 'Davies et al. (2007)', 'exon', row.start, row.end, '.', '+', '.', row['ID'], 'Davies-%s' % transcript_id]
            tbl += [f]
            # include CDS
            phase = 3 - (transcript_length % 3) if transcript_length % 3 else 0
            f = ['2L', 'Davies et al. (2007)', 'CDS', row.start, row.end, '.', '+', phase, row['ID'], 'Davies-%s' % transcript_id]
            tbl += [f]
    transcript_length += row['end'] - row['start'] + 1
tbl = etl.wrap(tbl)
tbl.totsv('davies_vgsc_model.txt')
tbl.addfield('attributes', lambda row: 'ID=%s;Parent=%s' % (quote(row.ID), quote(row.Parent))).cutout('ID', 'Parent').data().totsv('davies_vgsc_model.gff3')
tbl.displayall()
```
